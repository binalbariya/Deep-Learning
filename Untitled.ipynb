```python
from skimage.io import imread, imshow
import numpy as np
listt=[]
#Method1: Grayscale Pixel Values as Features
for i in range(4):
    
    # define filename
    filename = folder+"\\" +str(i) + '.png'
    # load image pixels
    image = imread(filename)
    print("SHAPE:"+str(image.shape))
    feature_matrix = np.zeros((120,120)) 
    for i in range(0,image.shape[0]):
        for j in range(0,image.shape[1]):
            feature_matrix[i][j] = ((int(image[i,j,0]) + int(image[i,j,1]) + int(image[i,j,2]))/3)
    features = np.reshape(feature_matrix, (120*120)) 
    print("Mean values:"+str(features.shape))
    listt.append([image.shape[:-1],features.shape])
    
    
```
